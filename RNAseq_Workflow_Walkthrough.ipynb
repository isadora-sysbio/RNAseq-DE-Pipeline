## FASTQ Files

RNA-seq analysis begins with FASTQ files, which contain raw sequencing reads generated by a sequencing instrument.

Each read contains:

* A sequence of nucleotides (A, T, C, G)
* A quality score for each nucleotide

Example:

@Read1

ATCGATCGATCG

*

IIIIIIIIIIII

The sequence represents the observed nucleotides, while the quality scores indicate confidence in each base call.

In this project, the raw sequencing data were stored as compressed FASTQ files:

* SRR390728_1.fastq.gz
* SRR390728_2.fastq.gz

## FastQC

FastQC performs quality control on sequencing reads before downstream analysis.

Key metrics include:

* Per-base sequence quality
* GC content
* Sequence duplication levels
* Adapter contamination
* Sequence length distribution

The purpose of FastQC is to identify potential issues in sequencing data before alignment.

For this project, FastQC reports were generated for both sequencing files and inspected using MultiQC.

## MultiQC

MultiQC combines results from multiple quality control reports into a single summary report.

Instead of opening individual FastQC reports one at a time, MultiQC aggregates all metrics into one interface.

In this project, MultiQC was used to review:

* Sequence quality
* GC content
* Duplication levels
* Adapter content
* Quality score distributions

This provided a high-level assessment of sequencing quality before alignment.

## Alignment

Alignment is the process of determining where sequencing reads originated in a reference genome.

In this project, HISAT2 was used as the aligner.

A useful analogy is GPS navigation:

* FASTQ reads are individual locations that need to be placed on a map.
* The reference genome is the map.
* Alignment determines where each read belongs.

The output of alignment is stored in a SAM file, which records the genomic location of each aligned read.

## SAM and BAM Files

SAM (Sequence Alignment Map) is a human-readable text format that stores aligned sequencing reads.

Each alignment record contains information such as:

* Read name
* Chromosome
* Genomic position
* Alignment quality
* CIGAR string

Because SAM files can become extremely large, they are commonly converted into BAM files.

BAM files contain the same information as SAM files but are stored in a compressed binary format that is more efficient for storage and analysis.

## BAM Sorting

After alignment, BAM files are commonly sorted by genomic position.

Sorting organizes alignments according to their coordinates in the reference genome.

This improves performance for downstream tools because reads from nearby genomic regions are stored together.

In this project:

* mini.bam was generated from the SAM file.
* mini_sorted.bam was generated using samtools sort.

Sorting does not change the biological information. It only reorganizes how the data are stored.

## BAM Indexing

A BAM index (.bai) allows software to quickly access specific regions of a BAM file without reading the entire file.

An analogy is the index of a textbook:

* BAM = the book
* BAI = the index at the back of the book

Instead of scanning every page, software can immediately jump to a desired genomic location.

In this project:

* mini_sorted.bam
* mini_sorted.bam.bai

were generated using Samtools.

## Gene Annotation (GTF)

A GTF file describes where genes are located on a genome.

For this educational example, a small annotation file was created containing two genes:

* TP53
* MYC

The annotation defines:

* Gene boundaries
* Coordinates
* Feature types

Without annotation information, aligned reads cannot be assigned to genes.

## Gene Quantification

Gene quantification converts aligned reads into gene-level expression measurements.

featureCounts was used to determine how many aligned reads overlapped each annotated gene.

Conceptually:

Read A → TP53

Read B → TP53

Read C → MYC

Results:

TP53 = 2 reads

MYC = 1 read

The goal is not to study individual reads but to summarize expression at the gene level.

## Count Matrix

The output of gene quantification is a count matrix.

A count matrix records how many sequencing reads were assigned to each gene.

Example:

Gene | Count

TP53 | 400

MYC | 700

In real RNA-seq studies, count matrices typically contain thousands of genes across many biological samples.

The count matrix serves as the input for differential expression analysis.

## Lessons Learned

Key concepts learned during this project:

* FASTQ stores raw sequencing reads.
* FastQC evaluates sequencing quality.
* MultiQC summarizes quality-control results.
* Alignment determines where reads originate in a reference genome.
* SAM and BAM store alignment information.
* BAM files are commonly sorted and indexed.
* GTF files define gene locations.
* Gene quantification converts aligned reads into gene-level counts.
* Count matrices are the foundation of differential expression analysis.

This project established the foundation required for larger transcriptomics studies using real human reference genomes and biological datasets.
